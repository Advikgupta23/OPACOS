```python
def calculate_conditional_pdf(age_values, feh_value, distance_value):
    # Evaluate the joint PDF for a grid of age values at the specific feh and distance values
    age_pdf_values = age_kde(age_values)
    
    # Joint PDF of feh and distance
    feh_distance_point = np.array([[feh_value], [distance_value]])
    joint_feh_distance_pdf = feh_distance_kde(feh_distance_point)[0]
    
    # Ensure joint_feh_distance_pdf is broadcastable with age_pdf_values
    joint_feh_distance_pdf = np.full_like(age_pdf_values, joint_feh_distance_pdf)
    
    # Debug prints to inspect values
    print(f"age_pdf_values: {age_pdf_values}")
    print(f"joint_feh_distance_pdf: {joint_feh_distance_pdf}")
    
    # Calculate joint PDF for age given specific feh and distance
    joint_age_feh_distance_pdf = age_pdf_values * joint_feh_distance_pdf
    
    # Marginal PDF by integrating over age (assuming age_values is a discrete set of points)
    marginal_feh_distance = np.sum(joint_age_feh_distance_pdf)
    
    # Debug print to inspect marginal PDF
    print(f"marginal_feh_distance: {marginal_feh_distance}")
    
    # Add a small epsilon to prevent division by zero
    epsilon = 1e-10
    marginal_feh_distance += epsilon
    
    # Calculate conditional PDF
    conditional_pdf = joint_age_feh_distance_pdf / marginal_feh_distance
    
    return conditional_pdf
```
